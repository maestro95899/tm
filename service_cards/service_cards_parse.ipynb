```python
def get_second(x):
    if pd.isnull(x):
        return ""
    elif len(x.split('/')) > 2:
        return x.split('/')[2]
    else:
        return ""
    
def get_third(x):
    if pd.isnull(x):
        return ""
    elif len(x.split('/')) > 3:
        return x.split('/')[3]
    elif x.isdigit():
        return x
    else:
        return ""
```

```python
docs = pd.DataFrame()
docs = data[(pd.notnull(data['name']) | pd.notnull(data['description'])) & (data.status == 3)]
docs['text'] = docs['name'].map(lambda x: x if pd.notnull(x) else "") + " " + docs['description'].map(lambda x: x if pd.notnull(x) else "")

docs['first'] = docs['parent_id'].map(lambda x: x.split('/')[1] if pd.notnull(x) else None)
docs['first'][pd.isnull(docs['first'])] = docs['category_id'][pd.isnull(docs['first'])].map(lambda x: x.split('/')[1] if pd.notnull(x) else "")

docs['second'] = docs['parent_id'].map(get_second)
docs['second'][docs['second'] == ""] = docs['category_id'][docs['second'] == ""].map(get_second)

docs['third'] = docs['parent_id'].map(get_third)
docs['third'][docs['third'] == ""] = docs['category_id'][docs['third'] == ""].map(get_third)
```
